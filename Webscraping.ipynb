# Predicting Vintage Rating of the Red Italian wine Brunello di Montalcino 
# Webscraping
In this notebook I scaped a page from WineSpector, the premier magazine on wine.  The goal was to create a dataset containing the rating of the wine vintages.  I later decided to use another source for vintage ratings. Please look at the 'code' notebook containing the entire project.

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# Journal's webpage
url = 'https://www.winespectator.com/vintage-charts/region/tuscany-brunello-di-montalcino'
res = requests.get(url)
res.status_code
```

```python
# Defining BeautifulSoup
soup = BeautifulSoup(res.content, 'lxml')
```

```python
table = soup.find('table')
table
```

```python
# Web Scraping
wine=[]

#for row in table.find('tbody').find_all('tr'):
for row in table.find_all('tr')[1:-1:2]:
    #print(row)
    #break
    item={}
    item['Vintage'] = row.find('a').text
    #break
    item['Score'] = row.find_all('td')[1].text[5:]
    item['Drink Window'] = row.find('span').text
    item['Description'] =  row.find_all('td')[3].text
    #food['carbs'] = list[4].text
    #food['restaurant'] = soup.find('h3').text.strip()
    wine.append(item)
    
vint = pd.DataFrame(wine)
```

```python
# Checking the dataset
vint.head()
```

```python
# Removing the word 'Description' from the beginning of the wine description
vint['Description'] = vint['Description'].str.replace('Description','')
```
